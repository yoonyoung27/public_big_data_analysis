###② 컬렉션 순회(ch6+)

```python
print('①. 리스트 순회 ')
for element in ['a','b','c','d']:
    print(element)

print('')
print('②. 튜플 순회')
for element in ('월','화','수'):
    print(element)

print('')
print('③. 레인지 순회')
for n in range(1,10,2):
    print(n)

print('')
print('④. 문자열 순회')
for character in '가나다라':
    print(character)

print('')
print('⑤. 집합 순회')
for element in {'사자','박쥐','늑대','곰'}:
    print(element)

print('')
print('⑥. 매핑 순회')
drinks = {
    '아메리카노':2500,
    '카페 라테':3000,
    '딸기 주스':3500
}
for k in drinks:
    print(k)

print('⑥*.')
for k, v in drinks.items():
    print(k)
    print(v)
```

* 집합에는 순서가 없어 변수에 어떤 순서로 원소가 대입될지 알 수 없다
* 매핑에는 요소가 사전에 입력된 순으로 키를 순회한다
* items() 메서드를 사용하여 키와 값을 나란히 순회할 수 있다

```python
seasons = ['봄', '여름', '가을', '겨울']
mountains = ['금강산', '봉래산', '풍악산', '개골산']
for season, mountain in zip(seasons, mountains):
    print(season + '에는 ' + mountain)
```

```python
board = [
         [['black', '룩'], None, None, None, None, None, None, None],
         [None, None, None, ['black', '킹'], None, None, None, None],
         [None, None, None, None, None, None, None, None],
         [None, None, None, None, None, None, None, None],
         [None, None, ['white', '비숍'], None, None, None, None, None],
         [None, None, None, None, None, None, None, None],
         [None, None, None, None, None, None, None, None],
         [None, None, None, None, ['white', '킹'], None, None, None],
]
for row in board:                 # 바깥쪽 리스트의 각 요소를 순회한다
    for piece in row:             # 안쪽 리스트의 각 요소를 순회한다
        if piece:                 # 요소가 있으면 I를, 없으면 .을 출력한다
            print('I', end=' ')
        else:
            print('.', end=' ')
    print()                       # 바깥쪽 리스트를 순회할 때마다 행을 바꾼다
```

####③.3 요소 정렬
* sorted() 함수 +a

```python
coll=[10,5,1,9,7,3]
for _ in coll:                                  # 거품 정렬
    for i in range(len(coll)-1):                # 한 단계 수행할 때마다 남아 있는 가장 큰 수가 제 자리로 이동한다
        if coll[i]>coll[i+1]:
            coll[i],coll[i+1]=coll[i+1],coll[i]
print(coll)

#sorted(coll) ; 위와 같은 식
```

```python
sorted(coll, reverse=True)

sorted([3, -5, 1, -2, -4], key=abs)

items = [
         {'name': '아메리카노', 'price': 2000},
         {'name': '카페 라테', 'price': 2500},
         {'name': '카푸치노', 'price': 2400},
]
sorted_items = sorted(items, key=lambda item:item['price'])
pprint.pprint(sorted_items)
```

* reverse 매개변수는 오름차순과 내림차순을 설정한다
* key 매개변수(함수)는 해당 함수에 값을 입력받아 변환한 값을 비교해 정렬한다

```python
def one_to_infinite():
    """1 - 무한대의 자연수를 순서대로 내는 생성기를 반환한다."""
    n = 1                            # n은 1에서 시작한다
    while True:                      # ❶ 무한 반복
        yield n                      # ❷ 실행을 일시정지하고 n을 반환한다
        n += 1                       # n에 1을 더한다
natural_numbers_1 = one_to_infinite()
```

```python
next(natural_numbers_1)
```
