####\* 문자열 연산

```python
text='아메'
print('1. +, *\n  ', text+'리카노',
      text*3, '\n')

print('2. 인덱싱\n  ', text[1], '\n')

print('3. 슬라이싱\n  ', text[0:1], '\n')

print('4. in, not in\n  ', '아' in text,
      '카' not in text)
```

####\* 문자열 함수, 메서드(전용 함수)

* __<len()>__ 문자열의 길이를 센다
* __<.index(text)>__ text가 문자열 안에서 처음 나오는 위치를 찾는다
* __<.join(text)>__ text 각각의 문자 사이에 문자열을 삽입한다


문자열 메서드.png

```python
book = ' 안나 카레니나, Leo Tolstoy '
poem = '''죽는 날까지 하늘을 우러러
한 점 부끄럼이 없기를,
잎새에 이는 바람에도
나는 괴로워했다.'''
print('1.',len(book),
      book.count('나'),
      book.find('카레'),
      book.index('카레'),
      book.rfind('나'))
print('2.',book.lower(),
      book.upper())
print('3.',book.replace(' ', '-'))
print('4.',book.strip(),
      book.lstrip(),
      book.rstrip())
print('5.',book.split(','),
      poem.splitlines(),
      '.'.join(book))
('1학년').isnumeric()
```

* split() 메서드는 리스트를 반환한다
* join() 메서드는 문자열로 이루어진 시퀀스에 사용할 수 있다

```python
print(book.find('카레라이스'))
book.index('카레라이스')
```

* find() 메서드는 찾는 문자열이 없을 때 -1을 반환한다
* index() 메서드는 찾는 문자열이 없을 때 오류가 발생한다
