```python
physics, chemistry, biology = 80, 90, 70
form = '물리학: {}, 화학: {}, 생물학: {}'
form.format(physics, chemistry, biology)
```

```python
'물리학: {0}, 생물학: {2}, 화학: {1}'.format(50, 60, 70)
'{0[0]}년 {0[1]}월 {0[2]}일'.format([2017, 11, 20])              # 시퀀스를 전달하는 경우
'{0[h]}시 {0[m]}분 {0[s]}초'.format({'h': 16, 'm': 30, 's': 0})  # 매핑을 전달하는 경우
import math
'원주율: {0.pi}'.format(math)                                    # 모듈의 경우
```

* 항목 번호, 인덱스, 키, 속성 변수를 이용하여 전달되는 순서에 상관없이 가리키는 값을 출력할 수 있다

```python
countries = [
             {'name': 'China', 'population': 1403500365},
             {'name': 'Japan', 'population': 126056362},
             {'name': 'South Korea', 'population': 51736224},
             {'name': 'Pitcairn Islands', 'population': 56}
]
form = '나라: {0:16} | 인구: {1:10}'
for country in countries:
    print(form.format(country['name'], country['population']))

print()
form = '나라: {0:16} | 인구: {1:010}'
for country in countries:
    print(form.format(country['name'], country['population']))

print()
form = '나라: {0:>16} | 인구: {1:>10}'
for country in countries:
    print(form.format(country['name'], country['population']))
```

* {항목번호:자리길이} 형식을 이용하여 데이터 출력 길이를 정할 수 있다
* 자리길이를 나타내는 수 앞에 0을 붙이면 공백 대신 0으로 빈자리를 채울 수 있다
* < 기호로 왼쪽 정렬을, > 기호로 오른쪽 정렬을 지정할 수 있다

###②.3 re 모듈
패턴을 기준으로 매치하는 텍스트를 찾고 조작  
* 메서드  
image.png

\* 『손에 잡히는 정규표현식』(벤 포터 저, 김경수 역, 인사이트) 참고

```python
import re
pattern_string = r'파이썬'
pattern = re.compile(pattern_string)

re.search(pattern, '파이프')                 # 매치하지 않는다 : None 반환
match = re.search(pattern, '즐거운 파이썬')  # 일부 텍스트와 매치한다 : Match 객체 반환
match.group()

re.sub(' ', '*', '즐거운 파이썬 프로그래밍')
re.split(pattern, '즐거운 파이썬 프로그래밍')
```

image.png
* 패턴과 매치한 텍스트는 Match 객체의 group() 메서드로 결과를 확인할 수 있다

```python
sample = '1789Python김파이fog'
pattern = re.compile(r'[가-힣]{2,4}')  # 두 자에서 네 자 사이인 한글 텍스트
match = re.search(pattern, sample)
match.group()

sample = '이름: 김파이, 연락처: 010-1234-5678, 주소: 부산 어딘가'
pattern = re.compile(r'01\d-\d{3,4}-\d{4}')  # 01X-XXXX-XXXX
match = re.search(pattern, sample)
match.group()
```

* 다양한 특수 기호로 패턴을 나타낼 수 있다

#####④.1.2.2 쓰기
open() 함수에서 파일을 여는 방식을 정하는 mode 변수의 값을 'w'로 지정한다
* print(..., file=file) 함수 : 텍스트 파일에 텍스트 데이터를 출력 (화면에 텍스트 데이터를 출력하는 것과 구별)
* file 객체의 write() 메서드

```python
print('당신의 이름은 무엇인가요?')
name = input()
with open('hello.txt', 'w') as file:
    print(name, '님 반가워요.', file=file)

from datetime import date
today = date.today()
text = f'안녕하세요! 오늘은 {today.month}월 {today.day}일 입니다.\n'
with open('hello.txt', 'w') as file:
    file.write(text)

from datetime import datetime
now = datetime.now()
text = f'현재 시각: {now.hour:02}:{now.minute:02}:{now.second:02}\n'
with open('time.txt', 'a') as file:
    file.write(text)

try:
    with open('한번만_기록.txt', 'x') as file:
        file.write('이 파일은 한 번만 작성됩니다.')
except FileExistsError:
    print('파일이 이미 존재합니다.')
```

* print() 함수는 전달된 인자가 문자열이 아니더라도 텍스트로 자동 변환해 주지만 write() 메서드에는 데이터를 문자열로 직접 변경해야 한다
* 'w' 모드로 파일을 열면 파일의 기존 내용은 삭제되므로, 기존 내용 뒤에 새롭게 덧붙여 쓰려면 'a' 모드로 파일을 연다
* 파일이 이미 존재하지 않을 때만 새로 작성하고 싶다면 'x' 모드로 파일을 연다 (파일이 이미 존재할 때에는 FileExistsError가 발생한다)

####④.2.1 csv 모듈
CSV : row를 텍스트의 행으로, column을 쉼표(,)로 구분된 항목으로 변환하는 방식을 이용하여 표 데이터를 직렬화한 형식

\* 파이썬에서는 표 데이터를 저장할 때 표의 각 행을 리스트로 담는 중첩 리스트가 사용된다

* csv.writer() 함수와 writerows() 메서드 : 표 형식의 데이터를 CSV 형식으로 직렬화
* csv.reader() 함수와 list() 함수 : CSV 형식의 데이터를 표 형식으로 해석

```python
import csv
table = [
    ['title', 'genre', 'year'],
    ['Interstella', 'SF', '2014'],
    ['Braveheart', 'Drama', '1995'],
    ['Mary Poppins', 'Fantasy', '1964'],
    ['Gloomy Sunday', 'Drama', '2000'],
    ]                                         # 표 데이터를 담은 중첩 리스트
with open('movies_output.csv', 'w') as file:  # movies_output.csv 파일을 쓰기 모드로 열기
    writer = csv.writer(file)                 # CSV 파일을 작성하는 쓰기 객체
    writer.writerows(table)                   # 표를 전체 행을 CSV 파일에 써넣는다

import csv
import pprint
with open('movies.csv') as file:
    reader = csv.reader(file)  # CSV 파일을 읽어들이는 읽기 객체
    movies = list(reader)      # CSV 파일 내용을 리스트로 읽어들인다
pprint.pprint(movies)          # 읽어들인 내용을 화면에 출력
```

* csv.writer() 함수는 전달받은 중첩 리스트를 파일 객체의 CSV 파일로 쓰는 쓰기 객체를 반환한다
* csv.reader() 함수는 전달받은 파일 객체의 CSV 파일을 중첩 리스트로 읽어들이는 읽기 객체를 반환한다

####④.2.2 json 모듈
* JSON
  * 숫자와 따옴표 데이터는 각각 파이썬의 수와 문자열에 대응된다
  * 중괄호 표현(객체)은 파이썬의 사전에 대응된다
  * 대괄호 표현(배열)은 파이썬의 리스트에 대응된다
  * 작은따옴표가 아니라 큰따옴표를 사용한다
  * 중괄호와 대괄호 안의 마지막 데이터 뒤에 콤마를 붙여서는 안 된다
  * 줄바꿈과 들여쓰기 등 공백 문자는 생략될 수 있다

\* CSV와 JSON  
CSV의 모든 값은 텍스트이지만, JSON은 수와 텍스트를 구별할 수 있다  
CSV는 열과 행만으로 데이터를 구조화하지만, JSON은 객체(사전과 유사)와 배열(리스트와 유사)을 이용해 데이터를 입체적으로 구조화할 수 있다.

* json.dumps() 함수 : 파이썬 컬렉션을 JSON으로 직렬화(문자열)
* json.loads() 함수 : JSON으로 직렬화된 텍스트를 파이썬 컬렉션으로 해석

```python
[
    {
        'title': 'Interstella',
        'genre': 'SF',
        'year': 2014,
        'starring': ['M. McConaughey', 'A. Hathaway', 'J. Chastain'],
    },
    {
        'title': 'Mary Poppins',
        'genre': 'Fantasy',
        'year': 1964,
        'starring': ['J. Andrews', 'D. Van Dyke'],
    },
]  # 표 형식

[
    {
        "title": "Interstella",
        "genre": "SF",
        "year": 2014,
        "starring": ["M. McConaughey", "A. Hathaway", "J. Chastain"]
    },
    {
        "title": "Mary Poppins",
        "genre": "Fantasy",
        "year": 1964,
        "starring": ["J. Andrews", "D. Van Dyke"]
    }
]  # JSON 형식
```

```python
import json                             # json 모듈 임포트
data = [
    {
        'title': 'Interstella',
        'genre': 'SF',
        'year': 2014,
        'starring': ['M. McConaughey', 'A. Hathaway', 'J. Chastain'],
    },
    {
        'title': 'Mary Poppins',
        'genre': 'Fantasy',
        'year': 1964,
        'starring': ['J. Andrews', 'D. Van Dyke'],
    },
    ]                                   # 직렬화하려는 데이터
json_data = json.dumps(data)            # 컬렉션을 JSON으로 직렬화
print(json_data)                        # 직렬화된 텍스트 확인
with open('movies.json', 'w') as file:  # 파일로 저장
    file.write(json_data)

import json                        # json 모듈 임포트
import pprint
with open('movies.json') as file:  # 텍스트 파일을 읽어들인다
    json_data = file.read()
data = json.loads(json_data)       # 읽어들인 텍스트 데이터를 역직렬화
pprint.pprint(data)                # 해석된 데이터 확인
```

```python
from pathlib import Path
Path('.')     # 현재 디렉터리의 경로
Path('C:/')   # 하드 디스크의 경로

path = Path('C:/python-programming/')  # 디렉터리 경로
path.exists()                          # 경로의 대상이 존재하는가?
path.is_file()                         # 경로의 대상이 파일인가?
path.is_dir()                          # 경로의 대상이 디렉터리인가?
path.parent                            # 한 단계 위 경로
path.name                              # 대상의 이름
path.suffix                            # 대상의 확장자
path.with_name('lisp-programming')     # 이름을 바꾼 경로
path.with_suffix('.backup')            # 확장자를 바꾼 경로
```

```python
def ls(path):
    """path 디렉터리에 포함된 모든 하위 항목을 화면에 출력한다."""
    path_obj = Path(path)            # 전달된 경로의 Path 객체 생성
    
    for item in path_obj.iterdir():  # 모든 하위 항목을 순회하며
        print(item)                  # 화면에 출력

ls('C:/')                            # C:\ 경로의 모든 하위 항목 출력

from pathlib import Path
path = Path('C:/')
path_list = list(path.iterdir())     # 디렉터리의 하위 항목 리스트
path_list
```

* iterdir() 메서드는 경로가 가리키는 디렉터리에 포함된 파일·디렉터리를 순회하는 반복자를 반환한다
* for 루프를 이용하여 모든 파일을 출력할 수도 있고, list를 이용하여 각 파일을 하나씩 출력할 수도 있다

```python
Path('dir1').mkdir()                     # dir1 디렉터리를 생성한다
Path('file1').touch()                    # file1 (빈) 파일을 생성한다
Path('dir2').replace(`dir1/dir2')        # dir2를 dir1 아래로 이동
Path('dir1/dir2').replace(Path('dir2'))  # dir2를 원위치로 이동
Path('dir2').replace('dir3')             # dir2의 이름을 dir3으로 변경
Path('dir3').rmdir()                     # dir3 디렉터리를 삭제한다
Path('file1').unlink()                   # file1 파일을 삭제한다

from pathlib import Path
def mv(src, dst):
    """src 경로의 파일을 dst 경로로 이동한다."""
    src_obj = Path(src)
    dst_obj = Path(dst)
    
    if dst_obj.exists():       # 대상 경로의 파일이 존재할 경우
        raise FileExistsError  # FileExistsError 예외를 일으키자
    src_obj.replace(dst_obj)

mv('old', 'new')
```

* replace() 메서드로 파일을 이동시키거나 이름을 수정할 때, 새로 지정한 경로에 이미 파일·디렉터리가 존재한다면 경로가 변경된 파일로 교체되어 버린다
