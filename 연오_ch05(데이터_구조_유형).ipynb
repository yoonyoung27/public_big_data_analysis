#####\* 리스트 연산

```python
number_list = [1,2,3,4,5]
alphabet_list = ['a','b','c','d','e']
list_list = [['a','b'],['c','d','e']]

print('1. +, *\n  ', number_list+alphabet_list,
      number_list*2, '\n')

print('2. 인덱싱\n  ', alphabet_list[1], '\n')

print('3. 슬라이싱\n  ', alphabet_list[1:3], '\n')

print('4. in, not in\n  ', 3 in number_list,
      0 not in number_list)
```

* __a:b__ a이상 b미만
* 연산은 시퀀스 내에 공통으로 적용된다

```python
print('1. 인덱싱+\n  ', alphabet_list[-1],
      alphabet_list[0],
      alphabet_list[-0],
      list_list[1][1], '\n')

print('2. 슬라이싱+\n  ', alphabet_list[::2],
      alphabet_list[1::2],
      alphabet_list[::-1], '\n')

number_list[1:2] = [200, 300]
print('3. 수정\n  ', number_list, '\n')

del number_list[3]
print('4. 삭제\n  ', number_list)
```

* __-__ 역으로
* __a::b__ a번째부터 b간격으로 슬라이싱

```python
copied_list_1 = alphabet_list[:]
copied_list_1[0] = 'A'
print(copied_list_1 == alphabet_list)

copied_list_2 = alphabet_list
copied_list_2[0] = 'A'
print(copied_list_2 == alphabet_list)
```

* copied_list_1은 다른 변수에 시퀀스를 복제한 것이고, copied_list_2는 같은 변수에 이름이 하나 더 생긴 것이다(ch8 참고)

#####\* 리스트 함수, 메서드(전용 함수)

* __<len()>__ 시퀀스의 길이를 센다
* __<.count(x)>__ 시퀀스 내에 x가 몇개 있는지 센다
* __<.sort()>__ 리스트의 요소를 순서대로 정렬한다
* __<.reverse()>__ 리스트를 역순으로 뒤집는다

시퀀스 메서드.png

#####* 통계 함수

* __<sum()>__ 모든 요소의 합을 구한다
* __<min()>__ 시퀀스에 포함된 최소 요소를 구한다
* __<max()>__ 시퀀스에 포함된 최대 요소를 구한다

```python
numbers = [30,20,40,10]

print('1.', len(numbers),
      numbers.count(20))

numbers.sort()
print('2.', numbers)

numbers.reverse()
print('3.', numbers)

numbers.append(100)
print('4.', numbers)

numbers.insert(4,50)
print('5.', numbers)

numbers.extend([101,102,103])
print('6.', numbers)

numbers += [104,105]
print('7.', numbers)

print('8.', numbers.pop(),
      numbers.pop(1))

numbers.remove(40)
print('9.', numbers)

numbers = numbers.clear()
print('10.', numbers)
```

* extend() 메서드는 += 연산으로 대신할 수 있다
* a.sort() %>% a.reverse()는 a.sort(reverse=True)와 같다
* pop(x) 메서드는 x번째 요소를 꺼낸다

```python
numbers = [30,20,40,10]
numbers.sort()
print(numbers)

numbers = [30,20,40,10]
sorted(numbers)
```

* a.sort() 메서드는 sorted() 함수(ch7 참고)로 대신할 수 있다

```python
numbers = [30,20,40,10]

print(sum(numbers),
      min(numbers),
      max(numbers))
```

```python
print(min(alphabet_list),
      max(alphabet_list))
print(min([1,2,3,'a']))
```

* sort(), min(), max() 함수를 사용하려면 시퀀스에 포함된 요소는 서로 비교할 수 있는 대상이어야 한다

####②.1 사전(dict)

```python
print({},
      {
          '아메리카노':2500,
          '카페 라떼':3000,
          '딸기 주스':3500
      },
      {1004:'value',(1,2,3):'value','key':'value'})
```

* 키에는 불변 데이터를 사용해야 한다

#####\* 사전 연산

```python
word_dict = {'cat':'고양이', 'hammer':'망치', 'rainbow':'무지개','book':'책'}
print(word_dict['cat'])
print('cat' in word_dict,
      'dog' not in word_dict)
```

```python
word_dict['cat'] = '야옹이'
word_dict['moon'] = '달'
word_dict
```

#####\* 사전 함수, 메서드(전용 함수)

* __<len()>__ 매핑의 길이를 센다
* __<.get(키)>__ 키에 해당하는 값을 구한다
* __<.update(매핑)>__ 매핑에 다른 매핑의 내용을 덮어쓴다
* __<.clear()>__ 매핑의 모든 요소를 삭제한다

\* 시퀀스와 사전 변환

* __<.keys()>__ 사전의 키 시퀀스를 얻는다
* __<.values()>__ 사전의 값 시퀀스를 얻는다
* __<.items()>__ 사전의 키-값 쌍 시퀀스를 얻는다
* __<dict(zip(키 시퀀스, 값 시퀀스))>__ 키 시퀀스, 값 시퀀스로부터 사전을 얻는다

```python
print('1.', len(word_dict),
      word_dict.get('dog'))

word_dict.update({'star':'별님','moon':'달님'})
print('2.', word_dict)

w_keys = word_dict.keys()
w_values = word_dict.values()
w_seq = word_dict.items()
print('3.', w_keys, w_values, w_seq)

print('4.', dict(zip(w_keys, w_values)))

word_dict.clear()
print('5.', word_dict)
```

* 인덱싱을 사용하면 해당 키가 없을 경우 오류가 발생하지만, get() 메서드를 사용하면 해당 키가 없을 경우 None을 반환한다
* 인덱싱을 사용하면 요소를 하나씩 추가해야 하지만, update() 메서드를 사용하면 여러 요소를 한번에 추가할 수 있다
* keys(), values(), items() 메서드를 사용한 결과물은 각각 dict_keys, dict_values, dict_items 데이터 유형을 갖는다. 따라서 시퀀스 함수를 사용하지 못하지만 for문에 사용된다

###③집합

집합.png

```python
print(set(),
      {'bakyeono@gmail.com', 'wldbsdudzz@gmail.com'},
      set(range(10)))
```

#####\* 집합 연산

```python
들짐승 = {'사자','박쥐','늑대','곰'}
날짐승 = {'독수리','매','박쥐'}
바다생물 = {'참치','상어','문어 괴물'}
print('늑대' in 들짐승,
      '곰' not in 들짐승)
```

#####\* 집합 함수, 메서드(전용 함수)

* __<len()>__ 집합의 길이를 센다
* __<.union()>__ 합집합 구하기
* __<.intersection()>__ 교집합 구하기
* __<.difference()>__ 차집합 구하기
* __<.symmetric_difference()>__ 대칭 차집합 구하기
* __<.issubset()>__ 부분집합인지 검사한다
* __<.isdisjoint()>__ 서로소인지 검사한다

집합 메서드.png

```python
print('1.', 들짐승.union(날짐승),
      들짐승|날짐승|{'인간'})

print('2.', 들짐승.intersection(바다생물),
      들짐승&날짐승)

print('3.', 들짐승.difference(날짐승),
      날짐승-들짐승)

print('4.', 들짐승.symmetric_difference(날짐승),
      들짐승^날짐승)

print('5.', 들짐승.issubset(날짐승),
      들짐승<=날짐승)

print('6.', 들짐승.isdisjoint(바다생물))

들짐승.add('인간')
print('7.', 들짐승)

들짐승.discard('인간')
print('8.', 들짐승)

들짐승.remove('곰')
print('9.', 들짐승)

print('10.', 들짐승.pop())

들짐승.clear()
print('11.', 들짐승)
```

* | 연산은 union() 메서드를 대신할 수 있으며 여러 집합을 연달아 합할 수 있다
* intersection() 메서드는 & 연산으로 대신할 수 있다
* difference() 메서드는 - 연산으로 대신할 수 있다
* symmetric_difference() 메서드는 ^ 연산으로 대신할 수 있다
* issubset() 메서드는 <= 연산으로 대신할 수 있다
